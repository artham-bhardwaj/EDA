## Step 1: Load the Zomato Dataset

```python
df = pd.read_csv('zomato.csv',encoding ='latin-1')
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

### Step 2: Find Missing Values

```python
df.isnull().sum()
```

```python
# Find list of all features which has null values
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
sns.heatmap(df.isnull(),yticklabels=False, cbar=False, cmap='viridis')
```

```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
# Merging Country with zomato dataset to find out countries too and saving in df_new
df_new= pd.merge(df, df_country, on='Country Code', how='left')
df_new['Country'].nunique()
df_country['Country'].nunique()
```

```python
df_new.Country.value_counts()
```

**Observation 1:** Zomato has most transaction/Customers in India and least in Canada

```python
# Creating pichart for the Same
# getting list of countries and there values in seperate lists
countries = df_new.Country.value_counts().index
countries
values = df_new.Country.value_counts().values
```

```python
# Plotting using countries and values list
plt.pie(values, labels=countries,autopct='%1.1f%%')
```

```python
# Can't undertand pie Chart for all countries so getting top 3 only
values = df_new.Country.value_counts().head(3).values
index = df_new.Country.value_counts().head(3).index
plt.pie(values, labels=index,autopct='%1.1f%%')
```

**Observation 2:** - India has the largest market Share of about 94.4%

### Step 3: Exploring more Numerical Varia

```python
# Finding all columns who have Integer Datatype
[feature for feature in df_new.columns if df_new[feature].dtype == 'int64']
```

```python
# Finding from which country most ratings are coming
rating_sum = df_new.groupby('Country')['Aggregate rating'].sum().sort_values(ascending=False)
```

```python
# Finding Best rating giving Country
rating_mean = df_new.groupby('Country')['Aggregate rating'].mean().sort_values(ascending=False)
```

```python
# Clubing mean and sum of ratings
sum_and_mean = pd.merge(rating_sum,rating_mean, on='Country',how='outer')
sum_and_mean.columns = ['Sum','Mean']
sum_and_mean.head()
```
